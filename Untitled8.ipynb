Assignment 1: Basic Matrix Operations
1. Create two 3Ã—3 matrices with random integers in Python.
2. Perform addition, subtraction, and multiplication operations on these matrices.
3. Compute the determinant, inverse, and rank of one of the matrices.

```python
A = np.random.randint(1, 10, size=(3, 3))
B = np.random.randint(1, 10, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

matrix_sum = A + B
matrix_diff = A - B
matrix_mul = A @ B

print("\nAddition (A + B):\n", matrix_sum)
print("\nSubtraction (A - B):\n", matrix_diff)
print("\nMatrix Multiplication (A @ B):\n", matrix_mul)
```

**(1) Generate and Plot Two Random Vectors:**

a. Create two vectors, each containing 15 random floats of your choice.

b. Plot these vectors on the same graph using matplotlib to compare how
they look.

```python
import numpy as np
import matplotlib.pyplot as plt

vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

print("Vector 1:\n", vector1)
print("Vector 2:\n", vector2)

plt.figure(figsize=(10, 5))
plt.plot(vector1, marker='o', label='Vector 1 ')
plt.plot(vector2, marker='x', label='Vector 2 ')

plt.title('Comparison of Two Random Vectors')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
```

**(2) Generate and Visualize a Random Matrix:**

a. Create a 4x4 matrix with random values of your choice.

b. Display the matrix using a heatmap with appropriate color mapping (e.g.,
seaborn or matplotlib).

c. Add labels to the heatmap to indicate row and column numbers.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

random_matrix = np.random.randint(1, 100, size=(4, 4))

plt.figure(figsize=(8, 6))
sns.heatmap(random_matrix, annot=True, cmap='autumn',
            xticklabels=[0, 1, 2, 3], yticklabels=[0, 1, 2, 3])

plt.title('4x4 Random Matrix Heatmap')
plt.xlabel('Column Number')
plt.ylabel('Row Number')

plt.show()

print("Generated 4x4 Matrix:\n", random_matrix)
```

**3. Matrix Operations and Visualization:**

a. Create two 4x4 matrices with random values of your choice.

b. Perform the following operations:

i) Matrix Addition

ii) Matrix Subtraction

iii) Matrix Multiplication (ensure matrices are compatible for
multiplication)

c. For each of these operations, visualize the result using bar plots. Each bar
should represent one element of the matrix.

```python
A = np.random.randint(0, 10, size=(3, 3))
B = np.random.randint(0, 10, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

matrix_sum = A + B
matrix_diff = A - B
matrix_prod = np.dot(A, B)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(range(1, 10), matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

axes[1].bar(range(1, 10), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

axes[2].bar(range(1, 10), matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

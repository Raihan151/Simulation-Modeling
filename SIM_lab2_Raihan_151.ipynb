## 1Ô∏è‚É£ Create Your Personal Matrix

Let:
- d1 = last digit of student ID
- d2 = second last digit of student ID

For this student:
- d1 = 2
- d2 = 4

Matrix A is defined as:

\[ A = \begin{bmatrix} d1 & d2 \\ d2 & d1 \end{bmatrix} \]

```
import numpy as np

A = np.array([[2, 4], [4, 2]])
A
```

## 2Ô∏è‚É£ Analyze Matrix A

```
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)
eig_A = np.linalg.eigvals(A)

print('Shape:', A.shape)
print('Determinant:', det_A)
print('Rank:', rank_A)
print('Eigenvalues:', eig_A)

if det_A != 0:
    inv_A = np.linalg.inv(A)
    print('Inverse of A:\n', inv_A)
else:
    print('Matrix A is not invertible')
```

## 3Ô∏è‚É£ Value Change Experiment

We create matrix B by adding +1 to one element of A.

**Changed element:** A[0][0] ‚Üí A[0][0] + 1

```
B = A.copy()
B[0, 0] += 1
B
```

## 4Ô∏è‚É£ Analyze Matrix B

```
det_B = np.linalg.det(B)
rank_B = np.linalg.matrix_rank(B)
eig_B = np.linalg.eigvals(B)

print('Shape:', B.shape)
print('Determinant:', det_B)
print('Rank:', rank_B)
print('Eigenvalues:', eig_B)

if det_B != 0:
    inv_B = np.linalg.inv(B)
    print('Inverse of B:\n', inv_B)
else:
    print('Matrix B is not invertible')
```

## 5Ô∏è‚É£ Comparison & Explanation

**1. Determinant Change:**
The determinant changes because modifying one value alters the matrix balance.

**2. Rank Change:**
The rank remains the same because the rows are still linearly independent.

**3. Eigenvalues:**
Eigenvalues change because they are sensitive to element values.

**4. Invertibility:**
Matrix B is slightly harder or easier to invert depending on how far the determinant moves from zero.

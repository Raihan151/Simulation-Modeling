```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample scalar and array inputs
scalar = 3.7
arr = np.array([10.48, 20.89, 30.54, 40.33, 50.62,60.12])

# Trigonometric functions
sin_values = np.sin(arr)
cos_values = np.cos(arr)
tan_values = np.tan(arr)
asin_values = np.arcsin(np.clip(arr / np.max(arr), -1, 1))  # Normalized and Clipping for valid range [-1, 1]
acos_values = np.arccos(np.clip(arr / np.max(arr), -1, 1))
atan_values = np.arctan(arr)

# Exponential and logarithm
exp_values = np.exp(arr)
log_values = np.log(arr)  # Natural logarithm

# Absolute value and square root
abs_values = np.abs(arr)
sqrt_values = np.sqrt(arr)

# Remainder when divided by a scalar (e.g., 2)
rem_values = np.remainder(arr, 2)

# Rounding functions
round_values = np.round(arr)
floor_values = np.floor(arr)
ceil_values = np.ceil(arr)

# Display results
print("sin:", sin_values)
print("cos:", cos_values)
print("tan:", tan_values)
print("asin:", asin_values)
print("acos:", acos_values)
print("atan:", atan_values)
print("exp:", exp_values)
print("log:", log_values)
print("abs:", abs_values)
print("sqrt:", sqrt_values)
print("rem:", rem_values)
print("round:", round_values)
print("floor:", floor_values)
print("ceil:", ceil_values)
```

```python
# 2Ã—3 matrix with float values
mat = np.array([[2.1, 4.7, 1.3],[5.9, 0.4, 3.8]])




print("Matrix:\n", mat)

# 1. Maximum and minimum values with indices
max_value = np.max(mat)
min_value = np.min(mat)

max_index = np.unravel_index(np.argmax(mat), mat.shape)
min_index = np.unravel_index(np.argmin(mat), mat.shape)



# 2. Vector length (total number of elements)
vec_length = mat.size


# 3. Sorting Row-wise descending sort
sorted_row_desc = np.sort(mat, axis=1)[:, ::-1]


# Entire matrix descending sort
sorted_entire_desc = np.sort(mat.flatten())[::-1]


# 4. Sum and Product
sum_values = np.sum(mat)
prod_values = np.prod(mat)


# 5. Median, Mean, Standard Deviation
median_value = np.median(mat)
mean_value = np.mean(mat)
std_dev = np.std(mat)
print("Matrix:\n", mat)
print("\nMax value:", max_value, "at index", max_index)
print("Min value:", min_value, "at index", min_index)
print("\nVector length:", vec_length)
print("\nRow-wise sorted (descending):\n", sorted_row_desc)
print("\nEntire matrix sorted (descending):", sorted_entire_desc)
print("\nSum of elements:", sum_values)
print("Product of elements:", prod_values)
print("\nMedian value:", median_value)
print("Mean value:", mean_value)
print("Standard deviation:", std_dev)
```
